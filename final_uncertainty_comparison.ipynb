```python
import os.path as path
import numpy as np
import Corrfunc
from Corrfunc.theory import wp
from Corrfunc.theory.DDrppi import DDrppi
from numpy import random
import matplotlib.pyplot as plt
import import_ipynb
from functions import plot_results, error_data
```

```python
def get_frac_err(stdev, meanwp):
    err = [stdev[i]/meanwp[i] for i in range (len(stdev))]
    return err

def plot_frac_err(rp_, frac_err, labels, colors, dotted = False, y_max = 0):
    for i in range(len(rp_)):
        if dotted:
            plt.plot(rp_[i], frac_err[i], color = colors[i], linestyle = '--')
        else:
            plt.plot(rp_[i], frac_err[i], color = colors[i])
    if y_max > 0:
        plt.ylim(0, y_max)
    plt.xscale('log')
    plt.xlabel('rpavg')
    plt.ylabel('fractional error (error/mean)')
    plt.title('Fractional error vs Mean rp_avg')
    plt.legend(labels)
```

# Mass > 10**14

```python
meanrp_14_100p_ss = []
frac_err_14_100p_ss = []

colors = ['blue', 'orange', 'green', 'red']
data_14_100p_0s = np.load('mass_e14.npz')
rpavg_14_100p_0s = data_14_100p_0s['array_one']
wp_14_100p_0s = data_14_100p_0s['array_two']
mean_rp_14_100p_0s, mean_wp_14_100p_0s, stdev_14_100p_0s = error_data(rpavg_14_100p_0s, wp_14_100p_0s)
```

```python
meanrp_14_100p_ss.append(mean_rp_14_100p_0s)
frac_err_14_100p_ss.append(get_frac_err(stdev_14_100p_0s, mean_wp_14_100p_0s))

scales = [30, 60, 100]
for scale in scales:
    data = np.load(f"uncertainty_14_{scale}.npz")
    rp_ = data['array_one']
    wp_ = data['array_two']
    mean_rp, mean_wp, stdev = error_data(rp_, wp_)
    frac_err_14_100p_ss.append(get_frac_err(stdev, mean_wp))
    meanrp_14_100p_ss.append(mean_rp)


plot_frac_err(meanrp_14_100p_ss, frac_err_14_100p_ss, ['scale = 0', 'scale = 30', 'scale = 60', 'scale = 100'], colors, y_max = 0.1)
plt.title('Fractional error vs Mean rp_avg: Mass > 10**14, variable scale')
```

```python
meanrp_14_pp_0s = []
frac_err_14_pp_0s = []

meanrp_14_pp_0s.append(mean_rp_14_100p_0s)
frac_err_14_pp_0s.append(get_frac_err(stdev_14_100p_0s, mean_wp_14_100p_0s))

percents = ['10', '1', '001']

for p in percents:
    data = np.load(f"subsamp_14_{p}_0.npz")
    rp_ = data['rp']
    wp_ = data['wp']
    mean_rp, mean_wp, stdev = error_data(rp_, wp_)
    frac_err_14_pp_0s.append(get_frac_err(stdev, mean_wp))
    meanrp_14_pp_0s.append(mean_rp)

plot_frac_err(meanrp_14_pp_0s, frac_err_14_pp_0s, ['100%', '10%', '1%', '0.1%'], colors, y_max = 0.1)
plt.title('Fractional error vs Mean rp_avg: Mass > 10**14, variable subsample')
```

```python
data_14_100p_30s = np.load('uncertainty_14_30.npz')
rpavg_14_100p_30s = data_14_100p_30s['array_one']
wp_14_100p_30s = data_14_100p_30s['array_two']
mean_rp_14_100p_30s, mean_wp_14_100p_30s, stdev_14_100p_30s = error_data(rpavg_14_100p_30s, wp_14_100p_30s)

meanrp_14_pp_30s = []
frac_err_14_pp_30s = []

meanrp_14_pp_30s.append(mean_rp_14_100p_30s)
frac_err_14_pp_30s.append(get_frac_err(stdev_14_100p_30s, mean_wp_14_100p_30s))

percents = ['10', '1', '001']
for p in percents:
    data = np.load(f"subsamp_14_{p}_30.npz")
    rp_ = data['rp']
    wp_ = data['wp']
    mean_rp, mean_wp, stdev = error_data(rp_, wp_)
    frac_err_14_pp_30s.append(get_frac_err(stdev, mean_wp))
    meanrp_14_pp_30s.append(mean_rp)

plot_frac_err(meanrp_14_pp_0s, frac_err_14_pp_0s, [], colors)
plot_frac_err(meanrp_14_pp_30s, frac_err_14_pp_30s, ['100%', '10%', '1%', '0.1%'], colors, dotted=True, y_max = 0.1)
plt.title('Fractional error vs Mean rp_avg: Mass > 10**13 (dotted line = scale of 30)')

```
