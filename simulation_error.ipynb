```python
import os.path as path
import numpy as np
import Corrfunc
from Corrfunc.theory import wp
from Corrfunc.theory.DDrppi import DDrppi
import matplotlib.pyplot as plt
import import_ipynb
import statistics
from functions import create_data, plot_results, wp_pairs_cross
```

```python
def error_data(rpavg_, wp_):
    wp_mean = []
    rpavg_mean = []
    stdev = []
    for i in range(20):
        all_wp = []
        all_rpavg = []
        for arr in range(20):
            all_wp.append(wp_[arr][i])
            all_rpavg.append(rpavg_[arr][i])
        wp_mean.append(statistics.mean(all_wp))
        rpavg_mean.append(statistics.mean(all_rpavg))
        stdev.append(statistics.stdev(all_wp))
    return rpavg_mean, wp_mean, stdev

def wp_vs_rpavg(rpavg_, wp_, yerr, loglog = True):
    plt.figure(figsize=(20, 15))
    plt.plot(rpavg_, wp_, color = 'lightblue')
    plt.errorbar(rpavg_, wp_, yerr, fmt = 'o', markersize = 12, capsize = 20, capthick = .4, color = 'mediumblue' , ecolor = 'mediumblue')
    plt.title("rpavg VS wp", fontsize = 35)
    plt.xlabel("rpavg", fontsize = 25)
    plt.ylabel("wp", fontsize = 25)
    if loglog:
        plt.loglog()

def error_v_rp(rpavg_, yerr, loglog = True):
    plt.plot(rpavg_, yerr)
    plt.title("rpavg VS error")
    plt.xlabel("rpavg")
    plt.ylabel("error (stdev)")
    if loglog:
        plt.loglog()
```

# Mass > 10^14 Error Calculations

```python
e14 = np.load('mass_e14.npz')
e14_rpavg = e14['array_one']
e14_wp = e14['array_two']
```

```python
plot_results(e14_rpavg, e14_wp)
print(len(e14_rpavg))
```

```python
rpavg_14_mean, wp_14_mean, wp_14_stdev = error_data(e14_rpavg, e14_wp)
wp_vs_rpavg(rpavg_14_mean, wp_14_mean, wp_14_stdev)
```

```python
wp_vs_rpavg(rpavg_14_mean, wp_14_mean, wp_14_stdev, loglog = False)
```

```python
error_v_rp(rpavg_14_mean, wp_14_stdev)
```

```python
error_v_rp(rpavg_14_mean, wp_14_stdev, loglog = False)
```
